```python
# Dependencies and Setup
import numpy as np
import pandas as pd

# Load data, read data files, and store in Pandas df
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

school_data_df = pd.DataFrame(school_data)

# Combine into single dataset
school_data_complete_df = pd.merge(student_data, school_data_df, how="left", on=["school_name", "school_name"])

print(school_data_complete_df)
```

```python
# Total number of schools, students, and budget

total_schools = school_data_df["school_name"].count()
total_students = school_data_df["size"].sum()
total_budget = school_data_df["budget"].sum()

# Average math score and reading score
avg_math_score = school_data_complete_df["math_score"].mean()
avg_reading_score = school_data_complete_df["reading_score"].mean()

# Percentage of students who passed math, reading, and both
passingmath = school_data_complete_df.query('math_score >70')["School ID"].count()/total_students*100
passingreading = school_data_complete_df.query('reading_score >70')["School ID"].count()/total_students*100
overallpassing = (avg_math_score + avg_reading_score)/2


DistrictSummary_df = {"Total Schools":[total_schools], "Total Students":[total_students], "Total Budget":[total_budget], 
                      "Average Math Score":[avg_math_score], "Average Reading Score":[avg_reading_score], 
                      "% Passing Math":[passingmath], "% Passing Reading":[passingreading], 
                      "% Overall Passing":[overallpassing]}


DistrictSummary_df = pd.DataFrame(DistrictSummary_df)
DistrictSummary_df.head()
```
